```python
import pandas as pd
import geopandas as gpd
import unicodedata
```

```python

df1 = pd.read_csv('tableExport (1).csv', encoding='UTF-8')
df2 = pd.read_csv('tableExport (2).csv', encoding='UTF-8')
# Concatenar os DataFrames
df_final = pd.concat([df1, df2])

df_final
```

```python
df_final = df_final.drop(columns=['Detalhes'])
```

```python
df_final.rename(columns={'Município': 'Municipio'}, inplace=True)
df_final.rename(columns={'Municípios Circunscritos': 'Municípios_Circunscritos'}, inplace=True)
df_final.rename(columns={'Denominação da Serventia': 'Denominação_Serventia'}, inplace=True)
df_final
```

```python
# corrigir nomes e codificção dos titulos
# Nomes das colunas
nomes_colunas = ['Municipio', 'Denominação_Serventia', 'Endereço', 'Telefone', 'Municípios_Circunscritos']

# Remover acentos e converter para minúsculas   
nomes_colunas_sem_acentos = [unicodedata.normalize('NFKD', coluna).encode('ASCII', 'ignore').decode('utf-8').lower() for coluna in nomes_colunas]

# Atribuir os novos nomes de colunas
df_final.columns = nomes_colunas_sem_acentos

# Exibir os novos nomes de colunas
print(df_final.columns)
```

```python
# Consultar as linhas que contêm a palavra 'TABELIONATO' na coluna 'Denominação da Serventia'
linhas_com_tabelionato = df_final[df_final['denominacao_serventia'].str.contains('TABELIONATO')]

# Mostrar a quantidade de linhas que contêm a palavra 'TABELIONATO'
quantidade_linhas_com_tabelionato = linhas_com_tabelionato.shape[0]
print("Quantidade de linhas com a palavra 'TABELIONATO':", quantidade_linhas_com_tabelionato)
print(linhas_com_tabelionato)
```

```python
#Carregar o shapefile dos municípios
shapefile_municipios = r"C:\Users\carlos_dmg\Documents\digitalmap-geo\SHAPES E MALHAS TERRITORIAIS\PI_Municipios_2021.shp"
municipios_gdf = gpd.read_file(shapefile_municipios,encoding = "UTF-8")
municipios_gdf
```

```python
# Junção dos dados
# Certifique-se de que as colunas de junção tenham os mesmos nomes ou converta-as para os mesmos nomes
municipios_com_cartorios = municipios_gdf.merge(df_final, how='left', left_on='NM_MUN', right_on='municipio')

# Passo 4: Salvar o novo shapefile
# municipios_com_cartorios.to_file("municipios_com_cartorios.shp")

```

```python
# agrupar os dados por 'municipios'
municipios_agrupados = df_final['municipios_circunscritos']
municipios_agrupados
```

```python
municipios_agregados = []
for index, row in df_final.iterrows():
    # Obter o nome do município principal e os municípios circunscritos
    municipio_principal = row['municipio']
    municipios_circunscritos = str(row['municipios_circunscritos']) if not pd.isna(row['municipios_circunscritos']) else ''
    municipios_circunscritos = municipios_circunscritos.split(', ')
    
    # Filtrar os polígonos dos municípios
    municipios_filtrados = municipios_com_cartorios[municipios_com_cartorios['NM_MUN'].isin([municipio_principal] + municipios_circunscritos)]
    
    # Unir as geometrias dos municípios
    geometria_agregada = municipios_filtrados.unary_union
    
    # Adicionar o nome do município principal e a geometria agregada à lista de municipios_agregados
    municipios_agregados.append({'municipio': municipio_principal, 'geometry': geometria_agregada})

# Passo 3: Criar um DataFrame GeoPandas com os municípios agregados
df_agregado = gpd.GeoDataFrame(municipios_agregados)

# Passo 4: Salvar o novo shapefile
# df_agregado.to_file("municipios_agregados.shp")
df_agregado
```

```python
# Mesclar os DataFrames com base no campo em comum 'municipio'
df_agregado_completo = pd.merge(df_agregado, df_final, on='municipio', how='left')
```

```python
# Remover linhas duplicadas
df_agregado_completo = df_agregado_completo.drop_duplicates()

```

```python
# Salvar o GeoDataFrame mesclado sem linhas duplicadas
df_agregado_completo = df_agregado_completo.set_crs(epsg=4674)
df_agregado_completo.to_file('serventias_extrajudiciais.shp')  # Salvar como um novo arquivo shapefile
```
